```python
#Adding columns 'x' , 'y' and 'z' together 
df['x_y_z'] = df['x'] + df['y'] + df['z']
print(df)
```

```python
#Drop column that was just added
df = df.drop(columns ='x_y_z', axis=1)
```

```python
#Print to check if column was dropped 
print(df)
```

```python
#Filter out to higher quality diamonds 
hq_df = df.loc[(df['cut']== 'Premium') & (df['price']>= 15000) & (df['carat']>=1.60)]
print(hq_df)
```

```python
#Create csv file containing higher quality diamonds 
hq_df.to_csv('highQuality.csv')
```
